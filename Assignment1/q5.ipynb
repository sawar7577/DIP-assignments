```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
im = cv.imread('./images/lena.jpg', cv.IMREAD_GRAYSCALE)
```

### Part 1
 > When MSB are kept at 0 the range of histogram is reduced to (0, 128) as opposed to (0, 256) earlier. The pixels of 2nd half i.e (128, 256) gets shifted to first half of the range i.e (0, 128) 

```python
_, plots = plt.subplots(1,2,figsize = (20,10))
plots[0].hist(im.flatten(), bins = np.arange(256))
plots[0].set_title('original histogram')
plots[1].hist(np.right_shift(np.left_shift(im.flatten(), 1), 1),  bins = np.arange(256))
plots[1].set_title('histogram with removal of MSB')
plt.show()
```

### Part 2
 > When the LSB is kept 0 the histogram does not changes much, only the odd values get reduced by 1. Only even intensities exist in the histogram 

```python
_, plots = plt.subplots(1,2,figsize = (20,10))
plots[0].hist(im.flatten(), bins = np.arange(256))
plots[0].set_title('original histogram')
plots[1].hist(np.left_shift(np.right_shift(im.flatten(), 1), 1),  bins = np.arange(256))
plots[1].set_title('histogram with removal of LSB')
plt.show()
```

### Part 3 : Baud Link
-  Number of pixels = $ 512*512 $
-  Number of bits per pixel = $ 1 + 1 + 8 = 10 $ (start bit, stop bit, pixel value)
-  $$ Time =  \frac{data \ to \ be \ transmitted}{data \ transmission \ rate} $$
- using 56K baud link, $$ time =  \frac{512*512*10}{56000} = 46.811s $$
- using 3000K baud link, $$ time =  \frac{512*512*10}{3000000} = 0.8738s $$
