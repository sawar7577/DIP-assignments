```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import numpy.fft as fft
```

```python
root = './DIP_2019_A3/'
```

```python
img = cv2.imread(root+'lena.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
def apply_filter(img, filt):
    fft_img = np.fft.fftshift(np.fft.fft2(img))
    filter_img = np.abs(np.fft.ifft2(np.fft.ifftshift(np.multiply(filt, fft_img))))
    return filter_img
```

```python
filt = np.array([
    [0,1,0],
    [1,2,1],
    [0,1,0]
])/6
```

```python
out = cv2.filter2D(img, -1, filt/6)
fft_img = fft.fftshift(fft.fft2(img))
_,plts = plt.subplots(1,2,figsize=(20,20))
plts[0].imshow(out, cmap='gray')
plts[0].set_title('original image')
plts[1].imshow(np.log(1+np.abs(fft_img)), cmap='gray')
plts[1].set_title('fourier of image')
plt.show()
```

```python
fft_filt = fft.fftshift(fft.fft2(filt, [img.shape[0], img.shape[0]]))
plt.imshow(np.log(1+np.abs(fft_filt)), cmap='gray')
plt.title('equivalent filter in frequency domain')
plt.show()
```

```python
out = apply_filter(img, fft_filt)
_, plts = plt.subplots(1,2,figsize=(20,20))
plts[0].imshow(img, cmap='gray')
plts[0].set_title('original image')
plts[1].imshow(out, cmap='gray')
plts[1].set_title('filtered image')
plt.show()
```
