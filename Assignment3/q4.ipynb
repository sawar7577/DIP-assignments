```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import numpy.fft as fft
```

```python
root = './DIP_2019_A3/'
```

```python
img = cv2.imread(root+'land.png', cv2.IMREAD_GRAYSCALE)
```

```python
fft_img = fft.fftshift(fft.fft2(img))
plt.imshow(np.log(1+np.abs(fft_img)), cmap='gray')
plt.show()
```

```python
fft_img[:90,125:135] = 0
fft_img[110:,125:135] = 0
fft_img[90:100,:120] = 0
fft_img[90:100,140:] = 0
fft_img[44:50,:] = 0
fft_img[:,76:82] = 0
fft_img[:,178:184] = 0
fft_img[146:152,:] = 0

plt.imshow(np.log(1+np.abs(fft_img)), cmap='gray')
plt.show()
```

```python
denoised_img = np.abs(fft.ifft2(fft.ifftshift(fft_img)))
_,plots = plt.subplots(1,2,figsize=(20,20))
plots[0].imshow(img, cmap = 'gray')
plots[1].imshow(denoised_img, cmap='gray')
plt.show()
```

### Reason
- Removing the high frequency noise that is present as white lines in the fft, denoises the image aptly
