```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import numpy.fft as fft
from scipy.signal import convolve2d
import time
```

```python
root = './DIP_2019_A3/'
```

```python
img = cv2.imread(root+'rectangle.jpg', cv2.IMREAD_GRAYSCALE)
```

### Part 1

```python
fft_img = fft.fftshift(fft.fft2(img))

img_rot = np.rot90(img)
fft_rot = fft.fftshift(fft.fft2(img_rot))

_,plts = plt.subplots(2,2,figsize=(20,20))
plts[1][0].imshow(np.abs(fft_img), cmap='gray')
plts[1][1].imshow(np.abs(fft_rot), cmap='gray')

plts[0][0].imshow(img, cmap='gray')
plts[0][1].imshow(img_rot, cmap='gray')
plt.show()
```

### Part 2

```python
fft_img = fft.fftshift(fft.fft2(img))

img_rol = np.roll(img, [105,105], [0,1])
fft_rol = fft.fftshift(fft.fft2(img_rol))

_,plts = plt.subplots(2,2,figsize=(20,20))
plts[1][0].imshow(np.abs(fft_img), cmap='gray')
plts[1][1].imshow(np.abs(fft_rol), cmap='gray')

plts[0][0].imshow(img, cmap='gray')
plts[0][1].imshow(img_rol, cmap='gray')
plt.show()
```
