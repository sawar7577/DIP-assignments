```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import numpy.fft as fft
from scipy.signal import convolve2d
import time
```

```python
root = './DIP_2019_A3/'
```

```python
img = cv2.imread(root+'bricks.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
def sampling(img, nx, ny):
    img = img[:,0::nx]
    img = img[0::ny,:]
    img = np.repeat(img, nx, axis=0)
    img = np.repeat(img, ny, axis=1)
    
    return img
```

```python
img2 = sampling(img, 2, 2)
img3 = sampling(img, 3, 3)
_,plts = plt.subplots(3,2,figsize=(20,20))
plts[0][0].imshow(img, cmap='gray')
plts[0][0].set_title('original image')
plts[0][1].imshow(np.log(1+np.abs(fft.fftshift(fft.fft2(img)))), cmap='gray')
plts[0][1].set_title('fft original image')
plts[1][0].imshow(img2, cmap='gray')
plts[1][0].set_title('image sampled with (0.5, 0.5)')
plts[1][1].imshow(np.log(1+np.abs(fft.fftshift(fft.fft2(img2)))), cmap='gray')
plts[1][1].set_title('fft')
plts[2][0].imshow(img3, cmap='gray')
plts[2][0].set_title('image sampled with (0.3, 0.3)')
plts[2][1].imshow(np.log(1+np.abs(fft.fftshift(fft.fft2(img3)))), cmap='gray')
plts[2][1].set_title('fft')
plt.show()

```

### Observations
- When nx = 0.5 and ny = 0.5, we can see that there is no information loss in the frequency domain, whereas when nx = 0.33 and ny = 0.33, we can see loss of information. Hence, the nyquist frequency is nx = 0.5 and ny = 0.5.
